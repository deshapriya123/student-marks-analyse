## Check for Missing Values

### Subtask:
Identify and visualize any missing values in the `df_marks` DataFrame to understand their extent and distribution.


**Reasoning**:
First, I will display the count of missing values for each column in the `df_marks` DataFrame to quantify the missing data. Then, I will visualize these missing values using a heatmap to provide a clear graphical representation of their distribution.



```python
print("Missing values before handling:")
print(df_marks.isnull().sum())

plt.figure(figsize=(10, 6))
sns.heatmap(df_marks.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
print(f"Number of duplicate rows before removal: {df_marks.duplicated().sum()}")
df_marks.drop_duplicates(inplace=True)
print(f"Number of duplicate rows after removal: {df_marks.duplicated().sum()}")
```

```python
df_marks.to_csv('df_marks_final.csv', index=False)
print('DataFrame exported to df_marks_final.csv')
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df_marks['Marks'], kde=True, bins=10)
plt.title('Distribution of Student Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
print("Descriptive Statistics for df_marks:")
display(df_marks.describe())
```

```python
print("Data Types of df_marks:")
display(df_marks.info())
```

```python
print("Distribution of Numerical Features:")

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df_marks['number_courses'], kde=True)
plt.title('Distribution of Number of Courses')

plt.subplot(1, 3, 2)
sns.histplot(df_marks['time_study'], kde=True)
plt.title('Distribution of Time Study')

plt.subplot(1, 3, 3)
sns.histplot(df_marks['Marks'], kde=True)
plt.title('Distribution of Marks')

plt.tight_layout()
plt.show()
```

```python
print("Correlation Matrix Heatmap:")

plt.figure(figsize=(8, 6))
sns.heatmap(df_marks.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of df_marks')
plt.show()
```

```python
print("Scatter Plots to visualize relationships:")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x='time_study', y='Marks', data=df_marks)
plt.title('Marks vs Time Study')

plt.subplot(1, 2, 2)
sns.scatterplot(x='number_courses', y='Marks', data=df_marks)
plt.title('Marks vs Number of Courses')

plt.tight_layout()
plt.show()
```
